# A - Задача на вычисление

Для пирога на 10 человек следует взять 1/10 фунта чернослива. Сколько граммов чернослива следует взять для пирога, рассчитанного на 3 человек? Считайте, что 1 фунт равен 0,4кг.

## Задание

Напишите программу на языке Python для решения задачи с использованием переменных.

```python
# Ваш код здесь
# 1. Входные данные
people = 10        # число человек, на которое рассчитан рецепт
prunes_per_10 = 1        # фунт чернослива на 10 человек
people_needed = 3        # число человек, на которое хотим приготовить
pound_to_kg = 0.4        # 1 фунт = 0.4 кг
kg_to_g = 1000           # 1 кг = 1000 г

# 2. Обработка данных
# Считаем, сколько фунтов нужно для 3 человек
prunes_needed_pounds = prunes_per_10 * (people_needed / people)
# 1 * (3/10) = 0.3 фунта

# Переводим в килограммы
prunes_needed_kg = prunes_needed_pounds * pound_to_kg
# 0.3 * 0.4 = 0.12 кг


# Переводим в граммы
prunes_needed_g = prunes_needed_kg * kg_to_g
# 0.12 * 1000 = 120 г

# 3. Вывод результатов
print("Нужно чернослива (граммы):", prunes_needed_g)
```

# B - Задача на ввод и вывод

Футболка стоила X рублей. После снижения цены она стала стоить Y рублей. На сколько процентов была снижена цена на футболку?

## Задание

Напишите программу на языке Python для решения задачи с использованием переменных.

```python
# Ваш код здесь
# 1. Входные данные. Здесь нужно определить переменные для хранения входных данных и ввести значения с клавиатуры.
X = float(input("Введите исходную цену футболки (X рублей): "))
Y = float(input("Введите новую цену футболки (Y рублей): "))

# 2. Обработка данных. Здесь нужно вычислить искомое значение и сохранить его в переменной.
discount_percent = ((X - Y) / X) * 100

# 3. Вывод результатов. Здесь нужно вывести вычисленное значение на экран.
print("Цена снизилась на", discount_percent, "%")
```

# C - Задача про Ибрагима

Ибрагим меняет цвет своего галстука по временам года. Летом он носит малиновый галстук, осенью - зеленый, зимой - красный, весной - черный.

## Задание

Напишите программу на языке Python, которая по введенному времени года выдаст цвет галстука для Ибрагима.

```python
# Ваш код здесь
# 1. Входные данные. Здесь нужно определить переменные для хранения входных данных и ввести значения с клавиатуры.
ties = {
    "лето": "малиновый",
    "осень": "зелёный",
    "зима": "красный",
    "весна": "чёрный"
}



# 2. Обработка данных. Здесь нужно вычислить искомое значение и сохранить его в переменной.
season = input("Введите время года: ").lower()
# 3. Вывод результатов. Здесь нужно вывести вычисленное значение на экран.
if season in ties:
    print("Ибрагим носит", ties[season], "галстук")
else:
    print("Неправильный ввод сезона")
```

# D - Задача на списки

Нужно сформировать список  lst из 20 элементов случайным образом. Элементы списка должны быть в диапазоне [-50;50]. Выведите элементы списка на экран.

```python
# Ваш код здесь
import random

# создаём список из 20 случайных чисел в диапазоне [-50, 50]
lst = [random.randint(-50, 50) for _ in range(20)]

# выводим элементы списка
print(lst)
```

#### Задание 1: Найдите длину списка

```python
# Ваш код здесь
length = len(lst)

print("Длина списка:", length)
```

#### Задание 2: Найдите максимальный и минимальный элемент списка

```python
# Ваш код здесь
# Находим максимальный элемент
max_element = max(lst)

# Находим минимальный элемент
min_element = min(lst)

print("Максимальный элемент:", max_element)
print("Минимальный элемент:", min_element)
```

#### Задание 3: Выведите список в обратном порядке

```python
# Ваш код здесь
print(lst[::-1])
```

#### Задание 4: Создайте копию списка в переменной mlst и отсортируйте список mlst по возрастанию

```python
# Ваш код здесь
# Создаём копию списка
mlst = lst.copy()

# Сортируем копию по возрастанию
mlst.sort()

print("Исходный список:", lst)
print("Отсортированный список:", mlst)
```

#### Задание 5: Дано число K=8: Поменяйте местами элементы списка с номерами 0 и K

```python
print (lst)
K = 8  # индекс для обмена

# Меняем местами элементы с индексами 0 и K
lst[0], lst[K] = lst[K], lst[0]

print(lst)
```

#### Задание 6: Выведите элементы списка, которые заканчиваются на 3

```python
# Ваш код здесь
for x in lst:
    if abs(x) % 10 == 3:
        print(x)
```

#### Задание 7: Найдите среднее арифметическое четных элементов списка

```python
# Ваш код здесь
# Находим чётные элементы
even_numbers = [x for x in lst if abs(x) % 2 == 0]

# Считаем среднее арифметическое
if len(even_numbers) > 0:
    average = sum(even_numbers) / len(even_numbers)
    print("Среднее арифметическое чётных элементов:", average)
else:
    print("Чётных элементов нет")
```

#### Задание 8: Пользователь вводит с клавиатуры символ. Его нужно сохранить в переменной С: Выведите все символы строки кроме этого символа

```python
# Ваш код здесь
# Ввод символа, который нужно исключить
C = input("Введите символ для удаления: ")

# Формируем новую строку без символа C
result = ""
for char in user_input:
    if char != C:
        result += char

# Выводим результат
print("Строка без символа", C, ":", result)
```

#### Задание 10: Найдите количество слов в строке, которые начинаются на b

```python
# Ваш код здесь
# Разбиваем строку на слова
words = user_input.split()

# Считаем количество слов, начинающихся на 'b'
count_b = 0
for word in words:
    if word.lower().startswith('b'):
        count_b += 1

print("Количество слов, начинающихся на 'b':", count_b)
```
